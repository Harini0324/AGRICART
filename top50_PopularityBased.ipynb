```python
def generate_farmer_name(used_names):
    first_names = ['John', 'Jane', 'Michael', 'Emily', 'David', 'Sarah', 'Daniel', 'Olivia', 'James', 'Sophia','Ramesh','Suresh','Mala','Kala','Ramya','Angel']
    last_names = ['Smith', 'Johnson', 'Williams', 'Jones', 'Brown', 'Davis', 'Miller', 'Wilson', 'Moore', 'Taylor']

    while True:
        name = f"{random.choice(first_names)} {random.choice(last_names)}"
        if name not in used_names:
            return name

# Create a list of 200 unique random farmer names
num_names = 100
used_names = set()
farmer_names = []

for _ in range(num_names):
    name = generate_farmer_name(used_names)
    farmer_names.append(name)
    used_names.add(name)

# Create a DataFrame
#df = pd.DataFrame({'farmer_name': farmer_names})

# Save the DataFrame to a CSV file
#df.to_csv('unique_farmer_names.csv', index=False)
```

```python
def generate_phone_number():
    return str(random.randint(9000000000, 9999999999))

df['phone_number'] = [generate_phone_number() for _ in range(len(df))]
```

```python
def generate_farmer_name1():
    return random.choice(farmer_names)

df['farmer_name'] = [generate_farmer_name1() for _ in range(len(df))]
```

# ## POPULARITY BASED : TOP 50  

```python
num_df=df.groupby('commodity').count()['modal_price'].reset_index()
num_df.rename(columns={'modal_price':'sold count'},inplace=True)
num_df
```

```python
avg_df=df.groupby('commodity').mean()['modal_price'].reset_index()
avg_df.rename(columns={'modal_price':'avg_price'},inplace=True)
avg_df
```

```python
popular_df=num_df.merge(avg_df , on='commodity')
popular_df
```

```python
popular_df = popular_df[ popular_df ['sold count']>15].sort_values('sold count',ascending=False).head(50)
popular_df
```

```python
popular_df.merge(df,on='commodity')
```
