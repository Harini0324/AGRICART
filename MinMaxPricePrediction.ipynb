```python
import pandas as pd
import random

# Load dataset
data = df

# Dictionary to store the minimum support price for each crop
min_msp = {}

# Iterate over the data
for index, row in data.iterrows():
    crop = row["CropType"]

    # If the crop is not in the dictionary, add it
    if crop not in min_msp:
        min_msp[crop] = row["Crop_price"]

    # Otherwise, update the minimum support price if necessary
    else:
        if row["Crop_price"] < min_msp[crop]:
            min_msp[crop] = row["Crop_price"]

# Add the minimum support price column to the dataframe
data["min_msp"] = data["CropType"].apply(lambda crop: min_msp[crop])

# Calculate the maximum support price for each crop

max_msp = {}
for crop in data["CropType"]:
    random_offset = random.randint(100, 150)
    max_msp[crop] = (data[data["CropType"] == crop]["Crop_price"] + random_offset).max()


data["max_msp"] = data["CropType"].apply(lambda crop: max_msp[crop])


# Add the maximum support price column to the dataframe
data["max_msp"] = data["CropType"].apply(lambda crop: max_msp[crop])

# Print the dataframe
print(data)

# Save the dataframe to a CSV file
data.to_csv("output.csv")
```
